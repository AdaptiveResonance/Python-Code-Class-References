```python
categorical_df=superstore[['StoreCode','StoreName']]
```

```python
categorical_df['store_count']=1
store_one_hot_encoded=pd.pivot_table(categorical_df, index=categorical_df.index,
                                    columns=['StoreName'],
                                    values=['store_count'])
```

```python
store_one_hot_encoded=store_one_hot_encoded.fillna(0)
```

```python
store_one_hot_encoded.columns=store_one_hot_encoded.columns.droplevel()
```

## Basic pivot provides the mean amount, not very useful

```python
pd.pivot_table(df, index=["name"], values=["amount"])
```

```python
pd.pivot_table(df, index=["name"], values=["amount"],aggfunc=np.sum)
```

```python
pd.pivot_table(df, index=["name"], values=["amount"],aggfunc=np.count_nonzero)
```

```python
pd.pivot_table(df, index=["company"], values=["amount"],aggfunc=np.sum)
```
